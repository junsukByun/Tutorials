# 주요 사항
## 0) train_path와 test_path를 입력할 때는 폴더/txt 두 가지 다 가능하다. 폴더를 입력하고 싶다면 '/home/storage/ssd/K0_SD_final/RPC'와 같은 식으로 입력해야 한다. 맨 마지막에 '/'는 붙이면 error가 발생하며 /RPC 의 하위 폴더는 반드시 Killer/NonKiller 두 가지 폴더가 존재해야 한다. 단 대소문자 문제가 번거로워서 killer, Killer, NonKiller, Nonkiller, nonKiller, nonkiller 어떤 식으로 저장이 되어 있어도 상관없이 모두 데이터를 불러올 수 있게 하였다.     학습 데이터는 99.jpg만 사용한다.
## 1) train할 때 write_data안에 변수 입력하지 않고 class 생성 시에 모든 변수 다 입력
## 2) train_data: valid_data 나누는 것 쉽게 만듬    ex) train_path랑 test_path가 똑같다면 train_data에서 자동으로 train_data:valid_data = 7:3으로 나눠서 학습, 다르면 다른상태로 학습
## 3) tfrecord_filename에는 TFrecord의 이름을 입력. 만약 이후에 똑같은 TFrecord로 학습을 이어서 하고 싶거나 변수를 변경하여 학습하고 싶다면 file_read_mode를 read로 하면 된다. 만약 기존의 TFrecord가 없다면 file_read_mode를  'write'로 해야 한다. 
## 4) TFrecord를 만들면 package안의 TFrecords 폴더 안에 tfrecord_filename으로 시작하는 train, test 의 record 파일이 저장된다.
## 5) model_nm을 입력하면 checkpoints안에 model_nm의 폴더가 생성되고 그 아래에 weight 각 epoch에 해당되는 iteration 번호별로 파일들이 생성된다. 이 중 가장 긴 이름의 파일에 학습된 weight 값들이 저장된다. 또한 checkpoints/checkpoint 안에는 model_nm-iteration(epoch별) 식으로 txt 형태로 저장 된다. 10 epoch을 학습하였으면 10개의 iteration 정보가 저장된다.
## 6) 학습 시에 똑같은 model_nm을 가지고 재학습하면 기존에 마지막으로 학습되었던 epoch 부터 이어서 학습된다. 또한 한 번 학습시에는 최대 100epoch 까지의 정보를 저장할 수 있게 코드를 변경했다. Make_Model.py의  saver = tf.train.Saver(max_to_keep=100)에서 변경하였으며 만약 max_to_keep을 따로 설정하지 않으면 5개의 epoch만 저장된다.
## 7) model을 생성하면 checkpoints/모델/ 안에 Modeling_information.txt 라는 파일이 생성된다. 열어보면 그 안에는 Modeling 할 때 setting 한 정보들이 담겨있다. 만약 기존 Model을 가지고 이어서 학습한 경우 그 밑에 start train again이 뜨며 새롭게 학습한 setting 정보들이 뜬다.

```python
def Prob_list_graph(self, bin_num=20, hist=True, norm_hist=False, custom_fname=None):
    NonKiller_prob = self.df_type.pred_1

    self.NK_prob = NonKiller_prob[np.where(self.true == 1)[0]]
    self.K_prob = NonKiller_prob[np.where(self.true == 0)[0]]

    if self.fatal_killer_model_yn == True:
        sns.distplot(self.NK_prob , hist=hist, color="skyblue", label="Nonfatal Killer + NonKiller(OK)", bins=bin_num, norm_hist=norm_hist, axlabel="prob_fatal")
        sns.distplot(self.K_prob , hist=hist, color="red", label="Fatal Killer", bins=bin_num, norm_hist=norm_hist, axlabel="prob_fatal")
    else:
        sns.distplot(self.NK_prob , hist=hist, color="skyblue", label="NonKiller(OK)", bins=bin_num, norm_hist=norm_hist, axlabel="prob_OK")
        sns.distplot(self.K_prob , hist=hist, color="red", label="Killer(NG)", bins=bin_num, norm_hist=norm_hist, axlabel="prob_OK")
    plt.title(self.model_name, y=1.02)
    plt.legend()
    if custom_fname:
        plt.savefig(custom_fname+'.png')
    else:
        plt.savefig(self.model_name+'.png')
    plt.show()
```
